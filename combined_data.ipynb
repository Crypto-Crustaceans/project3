```python
import pandas as pd
```

```python
df_blockchain = pd.read_csv("blockchain_data.csv")
df_market = pd.read_csv("market_data.csv")
df_social = pd.read_csv("social_data.csv")
```

```python
df1 = df_blockchain.merge(df_market, on="time", how="inner")
```

```python
df2 = df1.merge(df_social, on="time", how="inner")
```

```python
df3 = df2[["symbol", "time", "transaction_count", "large_transaction_count", "average_transaction_value", "high", "low", "open", "close", "volumeto", "reddit_comments_per_day"]]
```

```python
df3["total_avg_transaction_value"] = df3["transaction_count"] * df3["average_transaction_value"]
```

```python
df3['7_day_total_avg_transaction_value'] = df3.iloc[:,11].rolling(window=7).mean()
```

```python
df3['7_day_large_transaction_count'] = df3.iloc[:,3].rolling(window=7).mean()
```

```python
df3['7_day_reddit_comments_per_day'] = df3.iloc[:,10].rolling(window=7).mean()
```

```python
df3['7_day_volumeto'] = df3.iloc[:,9].rolling(window=7).mean()
```
