```python
import requests
import pandas as pd
from pprint import pprint
```

```python
# import api key
# get free api key from: https://min-api.cryptocompare.com/

from config import api_key
```

```python
query_url = f'https://min-api.cryptocompare.com/data/blockchain/histo/day?fsym=BTC&limit=2000&api_key={api_key}'
        
response = requests.get(query_url)
        
if response.status_code == 200:
    response_json = response.json()
```

```python
data = response_json['Data']['Data']
```

```python
df = pd.DataFrame(data)

df.head()
```

```python
df.to_csv('blockchain_data.csv', index=False)
```
