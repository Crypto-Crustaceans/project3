```python
import pandas as pd
```

```python
df_combined_data = pd.read_csv("combined_data.csv")
df_news_sentiment = pd.read_csv("news_sentiment.csv")
```

```python
df1 = df_combined_data.merge(df_news_sentiment, on="date", how="inner")
```

```python
df1.columns
```

```python
df1.tail()
```

```python
df2 = df1[['symbol', 'date', 'transaction_count', 'large_transaction_count',
       'average_transaction_value', 'high', 'low', 'open', 'close', 'volumeto',
       'reddit_comments_per_day', 'total_avg_transaction_value',
       '7_day_total_avg_transaction_value', '7_day_large_transaction_count',
       '7_day_reddit_comments_per_day', '7_day_volumeto', 'sentence',
       'updated_sentiment', 'sentiment_numeric']]
```

```python
df2['7_day_sentiment'] = df2.iloc[:,15].rolling(window=7).mean()
```

```python
df3 = df2.groupby(['date']).mean()
```

```python
df3.tail()
```

```python
df3.to_csv("combined_sentiment.csv", index=False)
```
