```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
import time
import datetime as dt
```

```python
# set up splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
article_list = []

# iterate through all pages
for i in range(729):
    
    # visit web page
    print(f'scraping page {i+1}...')
    page_url = f"https://news.bitcoin.com/page/{i+1}/?s=btc"
    browser.visit(page_url)
    time.sleep(1)
    
    # create html object and scrape into soup
    html = browser.html
    soup = bs(html, "html.parser")
    
    articles = soup.find_all("div", "item-details")

    for article in articles:
        
        # scrape title and date articles
        title = article.h3.a.attrs['title']
        date = article.div.span.time.attrs['datetime']
        
        #put variables in dictionary
        article_dict = {"title": title,
                        "date": date
                       }

        article_list.append(article_dict)

print('scraping complete!')
    
    
```

```python
df = pd.DataFrame(article_list)

df.head(50)
```

```python
titles_decoded = []
for title in list(df['title']):
    title = title.encode(encoding='cp1252', errors="ignore")
    title = title.decode(encoding='cp1252')
    title = title.encode(encoding='ascii', errors="ignore")
    title = title.decode(encoding='ascii')
    title = title.encode(encoding='utf8', errors="ignore")
    title = title.decode(encoding='utf8')
    titles_decoded.append(title)
```

```python
df['title'] = titles_decoded
```

```python
df.to_csv('btc_headlines.csv', index=False, encoding="utf-8")
```
